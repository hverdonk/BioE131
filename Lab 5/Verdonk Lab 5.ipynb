## Generating The Folded RNA Images

The program RNAfold (part of the Vienna software package) was used to generate the structures of the folded RNAs.

*For the sequences with all OBS unbound:* The program was run with only the command ```RNAfold```. When the terminal prompted for input, the RNA sequences were pasted in. The output was a postscript file that, when rendered in Apple Preview, yielded an image of the folded RNA structure.

*For the sequences with OBS unbound:* The program was run with the command ```RNAfold -C [Constraint File]```, where the constraint file was a text file containing the RNA sequence with dots under the unconstrained bases and Xs under those bound in the OBS region. The terminal did not prompt for imput. The output was a postscript file that, when rendered in Apple Preview, yielded an image of the folded RNA structure.

````python
#YES1
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

yes1_gen = mpimg.imread('YES-1_generated.png')
yesgenplot = plt.imshow(yes1_gen)
yesgentitle = plt.title("RNAfold Generated Structure of YES-1 (OBS1 Unbound)")
plt.show()

yes1 = mpimg.imread('YES-1.png')
yes1plot = plt.imshow(yes1)
yes1title = plt.title("Penchovsky & Breaker YES-1 RNA Structure (OBS1 Unbound)")
plt.show()

yes1_gen_OBS1 = mpimg.imread('YES1_OBS1_gen.png')
yes_gen_obs1_plot = plt.imshow(yes1_gen_OBS1)
yes_gen_obs1_title = plt.title("RNAfold Generated Structure of YES-1 (OBS1 Bound)")
plt.show()

yes1_OBS1 = mpimg.imread('YES1_OBS1.png')
yes_obs1_plot = plt.imshow(yes1_OBS1)
yes_obs1_title = plt.title("Penchovsky & Breaker YES-1 RNA Structure (OBS1 Bound)")
plt.show()

````

**Figure 1:**  
*OBS1 Unbound:* Comparing both RNA structures for YES-1, we can see that while both of them have one short stem and two hairpin loops that form a "T-shape", the structure generated by RNAfold has an additional loop, turning the "T-shape" into a conformation more similar to the + sign. However, the red regions are still unbound.

*OBS1 Bound:* Here, the RNA structure generated by RNAfold is much more similar to the conformation provided by Penchovsky & Breaker. The two red regions line up, indicating that this riboswitch would self-cleave.

````python
#NOT1
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

not1_gen = mpimg.imread('NOT-1_generated.png')
notgenplot = plt.imshow(not1_gen)
notgentitle = plt.title("RNAfold Generated Structure of NOT-1 (OBS1 Unbound)")
plt.show()

not1 = mpimg.imread('NOT-1.png')
not1plot = plt.imshow(not1)
not1title = plt.title("Penchovsky & Breaker NOT-1 RNA Structure (OBS1 Unbound)")
plt.show()

not1_obs1_gen = mpimg.imread('NOT1_OBS1_gen.png')
not1_obs1_plot = plt.imshow(not1_obs1_gen)
not1_obs1_title = plt.title("RNAfold Generated Structure of NOT-1 (OBS1 Bound)")
plt.show()

not1_obs1 = mpimg.imread('NOT1_OBS1.png')
not1_obs1_plot = plt.imshow(not1_obs1)
not1_obs1_title = plt.title("Penchovsky & Breaker NOT-1 RNA Structure (OBS1 Bound)")
plt.show()
````

*OBS1 Bound:* Both structures are similar, and the red regions are not aligned this time. This is what we would expect from a NOT gate with input TRUE.
